```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../data/dataset.csv')
respondents_dataset = pd.read_csv('../data/respondents-profile.csv', delimiter=';')
dataset = dataset.merge(right=respondents_dataset, how='left',left_on='respondent', right_on='id')
dataset.drop('id', axis=1, inplace=True)
```

```python
respondents = dataset.drop_duplicates(subset='respondent')
experiences = respondents[['solution','experience']].groupby('experience').count()
plt.pie(list(experiences['solution'].values), labels = list(experiences['solution'].keys()), autopct='%1.1f%%')
plt.rcParams['figure.figsize'] = (15,4)
plt.rcParams['font.size'] = 15
plt.savefig('../data/figures/dem-experience.pdf', bbox_inches='tight')
```

```python
respondents = dataset.drop_duplicates(subset='respondent')
role = respondents[['solution','role']].groupby('role').count()
plt.pie(list(role['solution'].values), labels = list(role['solution'].keys()), autopct='%1.1f%%', pctdistance=0.75)
plt.rcParams['figure.figsize'] = (15,4)
plt.rcParams['font.size'] = 14
plt.savefig('../data/figures/dem-role.pdf', bbox_inches='tight')
```

```python
respondents = dataset.drop_duplicates(subset='respondent')
education = respondents[['solution','education']].groupby('education').count()
plt.pie(list(education['solution'].values), labels = list(education['solution'].keys()), autopct='%1.1f%%')
plt.rcParams['figure.figsize'] = (15,4)
plt.rcParams['font.size'] = 15
plt.savefig('../data/figures/dem-education.pdf', bbox_inches='tight')
```
